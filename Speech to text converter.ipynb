```python
#speech to text converter from a given wav format audio file 

#import library
import speech_recognition as sr

# Initialize recognizer for recognizing the speech
r = sr.Recognizer()

# Reading the Audio file
# listening the audio file and store it in converted_text variable

with sr.AudioFile('male.wav') as source:
        r.adjust_for_ambient_noise(source)    # to remove ambient noise from the audio not 100% efficient but quite good

    converted_text = r.listen(source)
    
# recoginize_() method can throw an error if the API is unreachable, therefore we use exception handling
    try:
        
        # using google speech recognition
        text_file = r.recognize_google(converted_text)
        print('Converting the audio scripts into text..')
        print(text_file)
     
    except:
         print('Sorry, please run again...')
```

```python
# speech to text converter with default language = English, reading Microphone as source of voice

#import library

import speech_recognition as sr

# Initialize recognizer for recognizing the speech

r = sr.Recognizer()

# Reading Microphone as source
# listening the speech and store in converted_text variable

with sr.Microphone() as source:
    r.adjust_for_ambient_noise(source)    # to remove ambient noise from the audio not 100% efficient but quite good

    print("start talking now...")
    converted_text = r.listen(source)
    print("Time over, thanks")
# recoginize_() method can throw an error if the API is unreachable, therefore we use exception handling
    
    try:
        # using google speech recognition
        print(" Audio to text: "+r.recognize_google(converted_text))
    except:
         print("Sorry, I did not get that.. please try again")
```
